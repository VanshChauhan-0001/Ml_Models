```python
#Important libraries
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
#Fetching the dataset and splitting into traing and testing data
mnist = fetch_openml('mnist_784')

X,y= mnist['data'],mnist['target']
y = y.astype(np.int64)#this convert's int to string

X_train,X_test,y_train,y_test = X[:60000],X[60000:],y[:60000],y[60000:]
```

```python
#Here we created 2 new data holders which contain a 2 as True and non2 as False
y_train_two = (y_train==2)
y_test_two = (y_test == 2)
```

```python
#This is our main Logistic Regression Model
clf = LogisticRegression(tol=0.1)
clf.fit(X_train,y_train_two)
```

```python
#Here we get our predicted labels in Terms of True/False
y_predicted = clf.predict(X_test)
print(y_predicted)
```

```python
#Getting The Accuracy of Model
accuracy = accuracy_score(y_test,y_predicted)
print(f"Accuracy score of Model is: {accuracy * 10:.2f}%")
```

```python
#This is the best part of this model here you can provide your image
#This will check weather your image is a 2 or not 2
#The results will be in True or False
import numpy as np
from PIL import Image

def predict_image(image_data):

  image = Image.open(image_data).convert('L')
  image = image.resize((28, 28))
  image_array = np.array(image)
  image_array = np.invert(image_array)
  image_array = image_array.reshape(1, -1)

  prediction = clf.predict(image_array)
  return prediction[0]
```
